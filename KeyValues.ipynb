```python
import pandas as pd
```

```python
# From https://stackoverflow.com/questions/60154404/is-there-the-equivalent-of-to-markdown-to-read-data
df = pd.read_table("fingering-key-c.md", sep="|", header=0, index_col=1, skipinitialspace=True).dropna(axis=1, how='all').iloc[1:] 
df
```

```python
def  table_to_value(col):
    key_val = 0
    for ii, v in enumerate(col[2:].values):
        if v.strip().lower() == "x":
            key_val += 2**(7-ii)
    return key_val
```

```python
for name, col in df.iteritems():
    col[0] = table_to_value(col)
```

```python
print(df.to_markdown())
```
